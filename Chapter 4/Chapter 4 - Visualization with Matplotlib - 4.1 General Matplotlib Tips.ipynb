### show() or No show()? How to Display Your Plots

#### Plotting from a script

```python
# If using matplotlib within a script, plt.show() starts
# an event loop and checks for all currently active figure objects
# Save as myplot.py
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

# Can only be used once per Python session!
plt.show()
```

```python
# User can then run it from a terminal using python myplot.py
```

#### Plotting from an IPython notebook

```python
%matplotlib inline
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```
