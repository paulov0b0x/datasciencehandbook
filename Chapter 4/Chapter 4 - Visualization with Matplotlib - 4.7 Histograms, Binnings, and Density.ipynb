## Histograms, Binnings, and Density

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
plt.hist(data)
```

```python
# custom histogram
# ! normed now is replaced by the parameter density
plt.hist(data, bins=30, alpha=0.5,
        histtype='stepfilled', density=True, color='steelblue',
        edgecolor='none')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

# specifying custom parameters
kwargs = dict(histtype='stepfilled', density=True, alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# Counting the number of points in a given bin
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

### Two-Dimensional Histograms and Binnings

```python
# Created by dividing points among two-dimensional bins
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

#### plt.hist2d: Two-dimensional histogram

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
x, y
```

```python
# Counting points in two-dimensional bins
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
```

```python
xedges
```

#### plt.hexbin: Hexagonal bindings

```python
# Two-dimensional histogram creates a tessellation of squares
# across the axes. Another natural shape for such a tessellation
# is the regular hexagon, which can be used with the plt.hexbin
# routine.
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

#### Kernel density estimation

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
          origin='lower', aspect='auto',
          extent=[-3.5, 3.5, -6, 6],
          cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```
